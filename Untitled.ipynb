```python
import re
# Separar la columna "Location" utilizando la coma como delimitador
location_split = df_clean['Location'].str.split(',', expand=True)

# Asignar los valores separados a nuevas columnas en el DataFrame
for i in range(8):
    df_clean[f'Location_{i+1}'] = location_split[i]

# Función para limpiar y clasificar los valores
def classify_location(parts):
    country = ""
    state = ""
    plus_code = ""
    zone = ""
    city = ""

    for part in parts:
        if pd.isna(part):
            continue
        part = part.strip()
        if re.search(r'\bVenezuela\b', part, re.IGNORECASE):
            country = part
        elif re.search(r'\b(Lara|Miranda|Zulia|Caracas|Anzoátegui|Aragua|Bolívar)\b', part, re.IGNORECASE):
            state = part
        elif re.search(r'\b[2-9C-HJ-NP-TV-Z]{4}\+[2-9C-HJ-NP-TV-Z]{2,8}\b', part, re.IGNORECASE):
            plus_code = part
        elif city == "" and len(part.split()) == 1:
            city = part
        elif zone == "" and len(part.split()) > 1:
            zone = part

    return country, state, plus_code, city, zone

# Aplicar la función de clasificación a las columnas Location_1 a Location_8 y crear nuevas columnas
df_clean[['Pais', 'Estado', 'Plus Code', 'Ciudad', 'Zona']] = df_clean.apply(
    lambda row: pd.Series(classify_location(row[['Location_1', 'Location_2', 'Location_3', 'Location_4', 'Location_5', 'Location_6', 'Location_7', 'Location_8']])), axis=1
)

# Eliminar las columnas temporales de Location_1 a Location_8
df_clean.drop(columns=[f'Location_{i+1}' for i in range(8)], inplace=True)

# Mostrar las primeras filas del DataFrame para verificar
print(df_clean.head(30))
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt


# Lista predefinida de características binarias
binary_features = ['Alarma','Seguridad','Portón eléctrico','Con barrio cerrado','Acceso controlado','Parrillera','Piscina',
                  'Armarios','Baño de visitas','Terraza','Comedor','Vestier','Estudio','Estar','Patio','Suite','Balcón',
                  'Ático','Jardín','Cocina','Cuarto de servicio','Sala de juego','Con cuarto de lavado','Área para desayunar',
                  'Acceso a internet','Aire acondicionado','Calefacción','TV por cable','Línea telefónica',
                  'Gas natural','Planta eléctrica','Con energia solar','Con conexión para lavarropas','Agua',
                  'Cisterna','Calentador a gas/ eléctrico','Chimenea','Gimnasio','Jacuzzi','Estacionamiento para visitantes',
                  'Área de cine','Área de juegos infantiles','Con área verde','Ascensor','Cancha de básquetbol',
                  'Con cancha de fútbol','Cancha de paddle','Cancha de tenis','Con cancha polideportiva',
                  'Salón de fiestas','Sauna','Heladera','Amoblado']

# Separar el valor objetivo
y = df_model['Price']

# Seleccionar solo las características binarias
X_binary = df_model[binary_features]

# Escalar características binarias
scaler = StandardScaler()
X_binary_scaled = scaler.fit_transform(X_binary)

# Aplicar PCA sin especificar el número de componentes
pca = PCA()
X_pca = pca.fit_transform(X_binary_scaled)

# Graficar la varianza explicada acumulada
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulada')
plt.title('Gráfica de Codo de PCA')
plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% umbral de varianza explicada', color='red', fontsize=12)
plt.show()

# Seleccionar el número de componentes que explican al menos el 95% de la varianza
n_components = next(i for i, total_variance in enumerate(cumulative_explained_variance) if total_variance >= 0.95) + 1

print(f'Número de componentes que explican al menos el 95% de la varianza: {n_components}')

# Aplicar PCA con el número adecuado de componentes
pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_binary_scaled)

# Crear un nuevo dataframe con las componentes principales
df_pca = pd.DataFrame(X_pca, columns=[f'PC{i+1}' for i in range(n_components)])

# Integrar los componentes principales en el dataframe original (sin las características binarias originales)
df_non_binary = df_model.drop(columns=binary_features + ['Price'])
df_final = pd.concat([df_non_binary, df_pca], axis=1)
df_final['Price'] = y

print(df_final)
```
