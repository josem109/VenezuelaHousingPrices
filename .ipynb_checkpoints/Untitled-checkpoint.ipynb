```python
import pandas as pd

# Cargar el archivo de Excel
file_path = "master_state.xlsx"  # Asegúrate de que el archivo esté en la misma ruta que tu notebook
sheet_name = "DataReg"  # Nombre de la pestaña que quieres cargar

# Leer el contenido de la pestaña "Data" en un DataFrame
df = pd.read_excel(file_path, sheet_name=sheet_name)

# Mostrar el contenido del DataFrame
df.head()
df_clean = df
df_clean.drop(columns=['URL'], inplace=True)
```

```python
#Limpieza de la Data
df_clean['Meters'] = df_clean['Meters'].str.replace(' m² totales', '')
df_clean['Superficie total'] = df_clean['Superficie total'].str.replace(' m²', '')
df_clean['Superficie construida'] = df_clean['Superficie construida'].str.replace(' m²', '')
df_clean['Antigüedad'] = df_clean['Antigüedad'].str.replace(' años', '')
#Clean Numeric Fields
#df_clean['Price'] = df_clean['Price'].str.replace('.', '', regex=False)  # Eliminar los puntos
#df_clean['Price'] = df_clean['Price'].str.replace(',', '', regex=False)  # Eliminar las comas, si las hubiera
#df_clean['Price'] = df_clean['Price'].str.replace('[^\d.]', '', regex=True)  # Eliminar cualquier carácter no numérico
#df_clean['Price'] = pd.to_numeric(df_clean['Price'], errors='coerce')  # Convertir a numérico

df_clean.head()
```

```python
import re
# Separar la columna "Location" utilizando la coma como delimitador
location_split = df_clean['Location'].str.split(',', expand=True)

# Asignar los valores separados a nuevas columnas en el DataFrame
for i in range(8):
    df_clean[f'Location_{i+1}'] = location_split[i]

# Función para limpiar y clasificar los valores
def classify_location(parts):
    country = ""
    state = ""
    plus_code = ""
    zone = ""
    city = ""

    for part in parts:
        if pd.isna(part):
            continue
        part = part.strip()
        if re.search(r'\bVenezuela\b', part, re.IGNORECASE):
            country = part
        elif re.search(r'\b(Lara|Miranda|Zulia|Caracas|Anzoátegui|Aragua|Bolívar)\b', part, re.IGNORECASE):
            state = part
        elif re.search(r'\b[2-9C-HJ-NP-TV-Z]{4}\+[2-9C-HJ-NP-TV-Z]{2,8}\b', part, re.IGNORECASE):
            plus_code = part
        elif city == "" and len(part.split()) == 1:
            city = part
        elif zone == "" and len(part.split()) > 1:
            zone = part

    return country, state, plus_code, city, zone

# Aplicar la función de clasificación a las columnas Location_1 a Location_8 y crear nuevas columnas
df_clean[['Pais', 'Estado', 'Plus Code', 'Ciudad', 'Zona']] = df_clean.apply(
    lambda row: pd.Series(classify_location(row[['Location_1', 'Location_2', 'Location_3', 'Location_4', 'Location_5', 'Location_6', 'Location_7', 'Location_8']])), axis=1
)

# Eliminar las columnas temporales de Location_1 a Location_8
df_clean.drop(columns=[f'Location_{i+1}' for i in range(8)], inplace=True)

# Mostrar las primeras filas del DataFrame para verificar
print(df_clean.head(30))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Supongamos que df es tu DataFrame
# df = pd.read_csv('tu_archivo.csv') # Carga tu DataFrame

# Vamos a enfocarnos en las columnas relevantes
numeric_columns = ['Meters', 'Price', 'Rooms', 'Bathrooms', 'Floors', 'Grill', 'Garage',
                   'Private Area', 'Garden', 'Pool', 'Antiquity', 'Electric Gate', 'House Type',
                   'Internet', 'Air Conditioner', 'Cable', 'Phone', 'Gas', 'Electric Plant',
                   'Solar Energy']

# Convertir las columnas a numéricas (si no lo están)
for column in numeric_columns:
    df_clean[column] = pd.to_numeric(df_clean[column], errors='coerce')

# Método de visualización con Boxplot
plt.figure(figsize=(15, 10))
df.boxplot(column=numeric_columns)
plt.xticks(rotation=90)
plt.title('Boxplot para detección de outliers')
plt.show()

# Método de los Cuartiles
Q1 = df_clean[numeric_columns].quantile(0.25)
Q3 = df_clean[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Filtrar los outliers
outliers = ((df_clean[numeric_columns] < (Q1 - 1.5 * IQR)) | (df_clean[numeric_columns] > (Q3 + 1.5 * IQR)))
print("Outliers detectados por el método de los cuartiles:")
print(outliers.sum())

# Método del Z-Score
z_scores = np.abs(stats.zscore(df_clean[numeric_columns].dropna()))
outliers_z_score = (z_scores > 3)
print("Outliers detectados por el método del Z-Score:")
print(outliers_z_score.sum())
```

```python
import openpyxl
from openpyxl.styles import PatternFill
# Definir las columnas numéricas
numeric_columns = ['Meters', 'Price', 'Rooms', 'Bathrooms', 'Floors', 'Garage', 'Antiquity']  # Añade las columnas relevantes

# Calcular los cuartiles y el IQR
Q1 = df_clean[numeric_columns].quantile(0.25)
Q3 = df_clean[numeric_columns].quantile(0.75)
IQR = Q3 - Q1

# Identificar los outliers
outliers = df_clean[((df_clean[numeric_columns] < (Q1 - 1.5 * IQR)) | (df_clean[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]

# Crear un archivo Excel con los registros de outliers
outliers.to_excel('outliers.xlsx', index=False)

# Cargar el archivo Excel
wb = openpyxl.load_workbook('outliers.xlsx')
ws = wb.active

# Definir un relleno para resaltar las celdas
highlight = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")

# Resaltar las celdas que contienen outliers
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
    for cell in row:
        col_idx = cell.column - 1
        if ws.cell(row=1, column=cell.column).value in numeric_columns:
            original_value = df_clean.loc[cell.row - 2, ws.cell(row=1, column=cell.column).value]
            if cell.value != original_value:
                cell.fill = highlight

# Guardar el archivo Excel con las celdas resaltadas
wb.save('outliers_highlighted.xlsx')

print(f"Se han resaltado las celdas con outliers en 'outliers_highlighted.xlsx'")
```
